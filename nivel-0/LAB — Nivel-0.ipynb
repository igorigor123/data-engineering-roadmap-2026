## üü© FASE 2 ‚Äî Modelo Relacional (CURATED)

```python
%%bash
# Crear esquema CURATED
psql postgresql://root:root@localhost:5432/ny_taxi <<'SQL'
CREATE SCHEMA IF NOT EXISTS curated;
SQL
```

```python
%%bash
# Crear tabla con definici√≥n de tipos de datos correctos
psql postgresql://root:root@localhost:5432/ny_taxi <<'SQL'
DROP TABLE IF EXISTS curated.trip CASCADE;

CREATE TABLE curated.trip (
    trip_id BIGSERIAL PRIMARY KEY,
    vendor_id INTEGER,
    pickup_datetime TIMESTAMP,
    dropoff_datetime TIMESTAMP,
    passenger_count INTEGER,
    trip_distance NUMERIC,
    fare_amount NUMERIC,
    tip_amount NUMERIC,
    total_amount NUMERIC,
    pulocation_id INTEGER,
    dolocation_id INTEGER
);
SQL
```

```python
%%bash
# Inserci√≥n de registros desde RAW a CURATED (error pedag√≥gico al intentar insertar un text "1.0" en un campo integer)
psql postgresql://root:root@localhost:5432/ny_taxi <<'SQL'
INSERT INTO curated.trip (
    vendor_id,
    pickup_datetime,
    dropoff_datetime,
    passenger_count,
    trip_distance,
    fare_amount,
    tip_amount,
    total_amount,
    pulocation_id,
    dolocation_id
)
SELECT
    vendor_id::INTEGER,
    pickup_datetime::TIMESTAMP,
    dropoff_datetime::TIMESTAMP,
    passenger_count::INTEGER,
    trip_distance::NUMERIC,
    fare_amount::NUMERIC,
    tip_amount::NUMERIC,
    total_amount::NUMERIC,
    pulocation_id::INTEGER,
    dolocation_id::INTEGER
FROM raw.yellow_taxi_raw;
SQL
```

```python
%%sql
-- #EDA b√°sico al campo pickup_datetime (ac√° se usa JupySQL)
SELECT
    COUNT(*) AS total_trips,
    MIN(pickup_datetime),
    MAX(pickup_datetime)
FROM curated.trip;
```

```python
%%sql
-- Ingresos por d√≠a
SELECT
    DATE(pickup_datetime) AS day,
    SUM(total_amount) AS revenue
FROM curated.trip
GROUP BY day
ORDER BY day;
```

```python
%%sql
-- Creaci√≥n de indice (en capa curated)
CREATE INDEX idx_trip_pickup_datetime
ON curated.trip (pickup_datetime);
```

```python
%%sql
-- Explicaci√≥n de ejecuci√≥n de la consulta SQL (rendimiento)
EXPLAIN ANALYZE
SELECT *
FROM curated.trip
WHERE pickup_datetime >= '2024-01-15';
```

## üüß FASE 4 ‚Äî Preparaci√≥n para OLAP-Parte-2

```python
%%bash
# Creaci√≥n de tabla taxi_zone_lookup
psql postgresql://root:root@localhost:5432/ny_taxi <<'SQL'
CREATE TABLE IF NOT EXISTS raw.taxi_zone_lookup (
    location_id   INTEGER,
    borough       TEXT,
    zone_name     TEXT,
    service_zone  TEXT
);
```

```python
%%bash
# Cargar csv a tabla en RAW
psql postgresql://root:root@localhost:5432/ny_taxi <<'SQL'
\copy raw.taxi_zone_lookup FROM 'taxi_zone_lookup.csv'CSV HEADER;
SQL
```

```python
%%sql
-- Ac√° se usa JupySQL (esto requiere el primer script que hay en FASE 3 sobre conexi√≥n)
SELECT * FROM raw.taxi_zone_lookup
LIMIT 10
```

```python
%%sql
-- Crear vista ZONE solo con campos necesarios
CREATE OR REPLACE VIEW curated.zone AS
SELECT
    location_id,
    zone_name,
    borough
FROM raw.taxi_zone_lookup;
```

```python
%%sql
-- Consulta a la vista creada
SELECT * FROM curated.zone
LIMIT 10
```

```python
%%sql
-- Crea vista VENDOR a partir de curated.trip
CREATE OR REPLACE VIEW curated.vendor AS
SELECT DISTINCT
    vendor_id,
    CASE vendor_id
        WHEN 1 THEN 'Creative Mobile'
        WHEN 2 THEN 'VeriFone'
        ELSE 'Unknown'
    END AS vendor_name
FROM curated.trip;
```

```python
%%sql
-- Consulta a vista creada
SELECT * FROM curated.vendor order by 1
LIMIT 10
```

```python
%%bash
# Creaci√≥n de view vw_trip_denorm Vista denormalizada
psql postgresql://root:root@localhost:5432/ny_taxi <<'SQL'
CREATE OR REPLACE VIEW curated.vw_trip_denorm AS
SELECT
    t.trip_id,
    DATE(t.pickup_datetime) AS pickup_date,
    EXTRACT(HOUR FROM t.pickup_datetime) AS pickup_hour,
    t.trip_distance,
    t.total_amount,
    v.vendor_name,
    z.zone_name  AS pickup_zone,
    z.borough   AS pickup_borough
FROM curated.trip t
LEFT JOIN curated.vendor v
       ON t.vendor_id = v.vendor_id
LEFT JOIN curated.zone z
       ON t.pulocation_id = z.location_id;
```

```python
%%sql
-- Consulta a vista creada (se debe ejecutar conexi√≥n para cuando se vuelva a ejecutar el comando luego de volver a encender la maquina virtual) 
SELECT * FROM curated.vw_trip_denorm
LIMIT 10
```

```python
%%sql
-- Creaci√≥n de vista materializada. View Materialized mv_daily_zone(Pre-c√°lculo OLAP b√°sico)
CREATE MATERIALIZED VIEW curated.mv_daily_zone AS
SELECT
    DATE(pickup_datetime) AS pickup_date,
    pulocation_id,
    COUNT(*)              AS trips,
    SUM(total_amount)     AS revenue
FROM curated.trip
GROUP BY
    DATE(pickup_datetime),
    pulocation_id;
```

```python
%%sql
-- Consulta a tabla creada
SELECT * FROM curated.mv_daily_zone 
LIMIT 10
```
